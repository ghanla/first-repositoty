```python
import pandas as pd

def search_common(df, col1, col2):
    unique_values = df[col1].unique()
    
    result = []

    # 각 unique한 col1 값에 대해 col2에서의 관계 확인
    for val in unique_values:
        filtered_df = df[df[col1] == val]
        col2_counts = filtered_df[col2].value_counts().to_dict()
        
        result.append({
            col1: val,
            "col2_counts": col2_counts  # col2 값들의 빈도수 딕셔너리 형태로 저장
        })

    # 결과를 DataFrame으로 변환
    result_df = pd.DataFrame(result)

    return result_df

# 실행
result_df = search_common(df, 'cc_num', 'street' )


# 또는 Jupyter Notebook에서는 display() 사용
from IPython.display import display
display(result_df)

# col1과 co2의 관계를 count 하여 unique() 개수와 비교하는 과정을 통해 서로 같이 움직이는 데이터인지를 1차 확인.
```

```python

# distance 변화량 계산
df_data['distance_diff_prev'] = df_data['distance'].diff().abs()  # 이전 값과의 차이
df_data['distance_diff_next'] = df_data['distance'].diff(-1).abs()  # 이후 값과의 차이


# 시간 차이 계산 (이전 행과 비교) → timedelta 적용
df_data['time_diff'] = df_data['trans_date_trans_time'].diff()  # 시간 차이 계산 (datetime 타입 유지)

# 조건 설정: (이전 값 차이 크고 & 이후 값 차이 크고 & 시간 차이 > 30분)
df_data['flag'] = np.where(
    (df_data['distance_diff_prev'] > 0.7) &  # 이전 값과 거리 차이 큼
    (df_data['distance_diff_next'] > 0.7) &  # 이후 값과 거리 차이 큼
    (df_data['time_diff'] > pd.Timedelta("00:30:00")),  # 시간 차이가 30분 초과
    True, False
)
```

```python
df_data[df_data['flag'] == True] #발견. 이들의 is_fraud를 1로 변형.
```
